We want an interactive visualization of this data with:
1. choice of colormap
2. the option to take the log of the colormap

```python
# Step 1: colormap -- hard code a value
colormap = 'Reds'
plt.imshow(michigan, cmap=colormap)
plt.colorbar()
plt.show()
```

```python
# Step 2: make a function with the parameter as input
#colormap = 'Reds'
def make_mich_plot(colormap):
    # CMD + ]
    plt.imshow(michigan, cmap=colormap)
    plt.colorbar()
    plt.show()
```

```python
# Step 2.5 - test the function
make_mich_plot('Blues')
```

```python
plt.colormaps()
```

```python
# Step 3: decorate the function
@ipywidgets.interact(colormap=plt.colormaps())
def make_mich_plot(colormap):
    # CMD + ]
    plt.imshow(michigan, cmap=colormap)
    plt.colorbar()
    plt.show()
```

```python
my_colormap_list = ['terrain', 'magma', 'Reds']

@ipywidgets.interact(colormap=my_colormap_list)
def make_mich_plot(colormap):
    # CMD + ]
    plt.imshow(michigan, cmap=colormap)
    plt.clim(-352,352) # setting color limit (like xlim, ylim)
    plt.colorbar()
    plt.show()
```

```python
# Step 1 -- hardcode parameters
colormap = 'terrain'
take_log = True

if take_log:
    norm = matplotlib.colors.SymLogNorm(10) # base 10 log
else:
    norm = matplotlib.colors.Normalize() # default

# CMD + [
plt.imshow(michigan, cmap=colormap, norm=norm)
plt.clim(-352,352) # setting color limit (like xlim, ylim)
plt.colorbar()
plt.show()
```

```python
# Step 2 -- make parameters as input to a function
# CMD + /
# colormap = 'terrain'
# take_log = True

def make_mich_plot(colormap, take_log):
    if take_log:
        norm = matplotlib.colors.SymLogNorm(10) # base 10 log
    else:
        norm = matplotlib.colors.Normalize() # default
    
    # CMD + [
    plt.imshow(michigan, cmap=colormap, norm=norm)
    plt.clim(-352,352) # setting color limit (like xlim, ylim)
    plt.colorbar()
    plt.show()
```

```python
# Step 2.5 -- test
make_mich_plot('Blues', False)
```

```python
# Step 3 -- decorator
@ipywidgets.interact(colormap=my_colormap_list, take_log=True)
def make_mich_plot(colormap, take_log):
    if take_log:
        norm = matplotlib.colors.SymLogNorm(10) # base 10 log
    else:
        norm = matplotlib.colors.Normalize() # default
    
    # CMD + [
    plt.imshow(michigan, cmap=colormap, norm=norm)
    plt.clim(-352,352) # setting color limit (like xlim, ylim)
    plt.colorbar()
    plt.show()
```

```python
# depending on where we are coding, there might be an extra step
button = ipywidgets.Button(description='I am a button')

out = ipywidgets.Output() # capture the output of a function

def say_click(event): # event?
    with out:
        print('I have clicked!  Woo!')

button.on_click(say_click)
```

```python
my_obj.name, my_obj.age
```

```python
import traitlets
```

```python
# class MyObject():
#     name = 'unnamed' # string
#     age = 0 # integer

class MyTraitletObject(traitlets.HasTraits):
    name = traitlets.Unicode('unnamed') # string
    age = traitlets.Int(0) # integer
```

```python
my_t_obj = MyTraitletObject()
my_t_obj.age = 47
my_t_obj.name = 'Bob'
```

```python
print('Hello, my name is:', my_t_obj.name, 'and I am', my_t_obj.age, 'years old')
```

Pausing here -- using the traitlet library we can then look for changes in these traits ("event" gets explained!)
