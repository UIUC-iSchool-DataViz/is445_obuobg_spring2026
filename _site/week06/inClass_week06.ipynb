```python
# class MyObject():
#     name = 'unnamed'
#     age = 0

import traitlets
```

```python
class MyTraitletObject(traitlets.HasTraits):
    name = traitlets.Unicode('unnamed') # string
    age = traitlets.Int(0) # integer
```

```python
my_t_obj = MyTraitletObject()
```

```python
my_t_obj.name, my_t_obj.age
```

```python
def name_changed(change): # event? change?
    print(change)

my_t_obj.observe(name_changed, ['name'])
```

```python
my_t_obj.name = 'Linda'
```

```python
def name_changed2(change): # event? change?
    #print(change)
    print('Hello, my name is', change['new'], 'and I am', change['owner'].age, 'years old.')

my_t_obj.observe(name_changed2, ['name'])
```

```python
my_t_obj.name = 'Linda'
```

```python
my_t_obj.unobserve(name_changed2, ['name'])
```

```python
# 1. Data
x = np.arange(100)
y = np.random.random(100)

# 2. Scales
x_sc = bqplot.LinearScale()
y_sc = bqplot.LinearScale()

# 3. Axis
ax_x = bqplot.Axis(scale=x_sc, label='X Values')
ax_y = bqplot.Axis(scale=y_sc, label='Y Values', orientation='vertical')

# 4. Marks (sometimes we do axis before/after)
lines = bqplot.Lines(x=x, y=y, scales= {'x':x_sc, 'y':y_sc})

# 5. interactivity
pz = bqplot.interacts.PanZoom(scales={'x':[x_sc], 'y':[y_sc]})

# 6. put everything together into a figure
fig = bqplot.Figure(marks=[lines], axes=[ax_x, ax_y], interaction=pz)

fig
```

```python
# 1. Data
x = np.random.random(100)
y = np.random.random(100)

# 2. Scales
x_sc = bqplot.LinearScale()
y_sc = bqplot.LinearScale()

# 3. Axis
ax_x = bqplot.Axis(scale=x_sc, label='X Values')
ax_y = bqplot.Axis(scale=y_sc, label='Y Values', orientation='vertical')

# 4. Marks (sometimes we do axis before/after)
scatters = bqplot.Scatter(x=x, y=y, scales= {'x':x_sc, 'y':y_sc})

# 5. interactivity
selector = bqplot.interacts.FastIntervalSelector(scale=x_sc, marks=[scatters])

# 6. put everything together into a figure
fig = bqplot.Figure(marks=[scatters], axes=[ax_x, ax_y], interaction=selector)

fig
```

```python
# 1. data
data = np.random.random((10,10))

# 2. scales
col_sc = bqplot.ColorScale(scheme='Greens')
x_sc = bqplot.OrdinalScale() # try changing to LinearScale and see what happens
y_sc = bqplot.OrdinalScale()

# 3. axis
x_ax = bqplot.Axis(scale=x_sc, label='Column Bin')
y_ax = bqplot.Axis(scale=y_sc, label='Row Bin', orientation='vertical')
col_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')

# 4. mark
heatmap = bqplot.GridHeatMap(color=data, scales={'color':col_sc, 'row':y_sc, 'column':x_sc}, 
                            interactions={'click':'select'}, 
                            anchor_style={'fill':'magenta'})

# 5. interactivity
def on_selected(change):
    #print(change)
    if len(change['owner'].selected) == 1: # only one single selection at a time
        #print(change['owner'].selected)
        i,j = change['owner'].selected[0] # check if i,j should be j,i by click on some values
        myLabel.value = 'Data value = ' + str(data[i,j])

heatmap.observe(on_selected, 'selected')

# 6. all together into a figure
fig = bqplot.Figure(marks=[heatmap], axes=[col_ax, x_ax, y_ax])
#fig
dashboard = ipywidgets.VBox([myLabel,fig])
dashboard
```

## Adding complexity -- 3D heatmap with random data

```python
data3d = np.random.random((10,10,20))
```

```python
data3d.shape
```

```python
# 1. data

# 2. scales
col_sc = bqplot.ColorScale(scheme='Greens')
x_sc = bqplot.OrdinalScale() # try changing to LinearScale and see what happens
y_sc = bqplot.OrdinalScale()

# 3. axis
x_ax = bqplot.Axis(scale=x_sc, label='Column Bin')
y_ax = bqplot.Axis(scale=y_sc, label='Row Bin', orientation='vertical')
col_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')

# 4. mark
heatmap = bqplot.GridHeatMap(color=np.sum(data3d,axis=2), # changed this for 3d data
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc}, 
                            interactions={'click':'select'}, 
                            anchor_style={'fill':'magenta'})

# 5. interactivity
myLabel = ipywidgets.Label()
def on_selected(change):
    if len(change['owner'].selected) == 1: # only one single selection at a time
        i,j = change['owner'].selected[0] # check if i,j should be j,i by click on some values
        myLabel.value = 'Data value (sum) = ' + str(np.sum(data3d[i,j,:])) 

heatmap.observe(on_selected, 'selected')

# 6. all together into a figure
fig = bqplot.Figure(marks=[heatmap], axes=[col_ax, x_ax, y_ax])
#fig
dashboard = ipywidgets.VBox([myLabel,fig])
dashboard
```

```python
heatmap.selected
```

```python
i,j = heatmap.selected[0]
i,j
```

```python
data3d[i,j,:] # 1. data for histogram
```

```python
# 2. scales
x_sch = bqplot.LinearScale()
y_sch = bqplot.LinearScale()

# 3. axes for histogram
x_axh = bqplot.Axis(scale=x_sch, label='Value for 3rd axis')
y_axh = bqplot.Axis(scale=y_sch, label='Frequency', orientation='vertical')
```

```python
# 4. mark
#bqplot.Hist?
hist = bqplot.Hist(sample=data3d[i,j,:], scales={'sample':x_sch, 'count':y_sch}, bins=5)
```

```python
# 5. interactivity for histogram (number of bins change?)

# 6. figure
figh = bqplot.Figure(marks=[hist], axes=[x_axh, y_axh])
figh
```

```python
# 1. data

# 2. scales
col_sc = bqplot.ColorScale(scheme='Greens')
x_sc = bqplot.OrdinalScale() # try changing to LinearScale and see what happens
y_sc = bqplot.OrdinalScale()

# 3. axis
x_ax = bqplot.Axis(scale=x_sc, label='Column Bin')
y_ax = bqplot.Axis(scale=y_sc, label='Row Bin', orientation='vertical')
col_ax = bqplot.ColorAxis(scale=col_sc, orientation='vertical', side='right')

# 4. mark
heatmap = bqplot.GridHeatMap(color=np.sum(data3d,axis=2), # changed this for 3d data
                             scales={'color':col_sc, 'row':y_sc, 'column':x_sc}, 
                            interactions={'click':'select'}, 
                            anchor_style={'fill':'magenta'})

# 5. interactivity
myLabel = ipywidgets.Label()
def on_selected(change):
    if len(change['owner'].selected) == 1: # only one single selection at a time
        i,j = change['owner'].selected[0] # check if i,j should be j,i by click on some values
        myLabel.value = 'Data value (sum) = ' + str(np.sum(data3d[i,j,:])) # updating the value attribute of myLabel
        hist.sample = data3d[i,j,:] # update the sample attribute of hist 

heatmap.observe(on_selected, 'selected')

# 6. all together into a figure
fig = bqplot.Figure(marks=[heatmap], axes=[col_ax, x_ax, y_ax])
#fig
#dashboard = ipywidgets.VBox([myLabel,fig])
figures = ipywidgets.HBox([fig, figh])
dashboard = ipywidgets.VBox([myLabel, figures])
fig.layout.min_width = '400px'
figh.layout.min_width = '400px'
dashboard
```
