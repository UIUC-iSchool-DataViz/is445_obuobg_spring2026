```python
fig, ax = plt.subplots(figsize=(10,2))
ax.hist(im_data[:,:,0].flatten(), bins=256, color='red') # trying to plot just red channel
ax.hist(im_data[:,:,1].flatten(), bins=256, color='green') # trying to plot just green channel
ax.set_xlabel('Red channel bins')
ax.set_ylabel('# of pixels per bin')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.hist(im_data[:,:,0].flatten(), bins=256, color='red') # trying to plot just red channel
ax.hist(im_data[:,:,1].flatten(), bins=256, color='green') # trying to plot just green channel
ax.hist(im_data[:,:,2].flatten(), bins=256, color='blue') # trying to plot just blue channel
ax.hist(im_data[:,:,3].flatten(), bins=256, color='gray') # trying to plot just alpha/opacity channel
ax.set_xlabel('Red channel bins')
ax.set_ylabel('# of pixels per bin')
plt.show()
```

```python
fig, ax = plt.subplots(4, 1, figsize=(10,8))
ax[0].hist(im_data[:,:,0].flatten(), bins=256, color='red') # trying to plot just red channel
ax[1].hist(im_data[:,:,1].flatten(), bins=256, color='green') # trying to plot just green channel
ax[2].hist(im_data[:,:,2].flatten(), bins=256, color='blue') # trying to plot just blue channel
ax[3].hist(im_data[:,:,3].flatten(), bins=256, color='gray') # trying to plot just alpha/opacity channel
# updating x-axis labels
ax[0].set_xlabel('Red channel bins')
ax[1].set_xlabel('Green channel bins')
ax[2].set_xlabel('Blue channel bins')
ax[3].set_xlabel('Alpha channel bins')

for a in ax:
  a.set_ylabel('# of pixels per bin')
fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(5, 1, figsize=(10,8))
ax[0].hist(im_data[:,:,0].flatten(), bins=256, color='red') # trying to plot just red channel
ax[1].hist(im_data[:,:,1].flatten(), bins=256, color='green') # trying to plot just green channel
ax[2].hist(im_data[:,:,2].flatten(), bins=256, color='blue') # trying to plot just blue channel
ax[3].hist(im_data[:,:,3].flatten(), bins=256, color='gray') # trying to plot just alpha/opacity channel
# updating x-axis labels
ax[0].set_xlabel('Red channel bins')
ax[1].set_xlabel('Green channel bins')
ax[2].set_xlabel('Blue channel bins')
ax[3].set_xlabel('Alpha channel bins')

for a in ax[:3]:
  a.set_ylabel('# of pixels per bin')

ax[4].imshow(im_data) # plotting the image with "imshow", using the *data* of the image
fig.tight_layout()
plt.show()
```
