```python
buildings = pd.read_csv('https://github.com/UIUC-iSchool-DataViz/is445_data/raw/main/building_inventory.csv', 
                       na_values = {
                           'Square Footage':0, 
                           'Year Constructed':0, 
                           'Year Acquired':0
                       })
```

```python
buildings['Square Footage'].min()
```

```python
buildings['Year Constructed'].min(), buildings['Year Acquired'].min()
```

```python
buildings['Square Footage'].plot()
```

```python
buildings['Square Footage'].min(), buildings['Square Footage'].max()
```

```python
ax = buildings.plot(x='Year Acquired', y='Square Footage', kind='scatter')
ax.set_yscale('log')
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(10,2))
#ax = 
buildings.plot(x='Year Acquired', y='Square Footage', kind='scatter', ax=ax)
ax.set_yscale('log')
```

```python
buildings['Bldg Status'].unique()
```

```python
for status, group in buildings.groupby('Bldg Status'):
    print(status)
    print(group.shape)
```

```python
for status, group in buildings.groupby('Year Acquired'):
    print(status)
    print(group.shape)
```

```python
buildings.groupby('Year Acquired')['Square Footage'].sum()
```

```python
fig, ax = plt.subplots(figsize=(10,2))
gbs = buildings.groupby('Year Acquired')['Square Footage'].sum()
gbs.plot(ax=ax) # using pandas that uses matplotlib to plot
ax.set_yscale('log')
ax.set_ylabel('Total Square Footage')
plt.show()
```

```python
x = gbs.index.values
y = gbs.values

fig, ax = plt.subplots(figsize=(10,2))
gbs = buildings.groupby('Year Acquired')['Square Footage'].sum()
#gbs.plot(ax=ax) # using pandas that uses matplotlib to plot
ax.plot(x,y) # using matplotlib directly
ax.set_yscale('log')
ax.set_ylabel('Total Square Footage')
ax.set_xlabel('Year Aquired')
plt.show()
```

```python
stats = buildings.groupby('Year Acquired')['Square Footage'].describe()
stats
```

```python
fig, ax = plt.subplots(figsize=(10,2))
ax.plot(stats['max'], marker='.')
ax.set_yscale('log')
ax.set_ylabel('Square Footage stats')
ax.set_xlabel('Year Acquired')
plt.show()
```
